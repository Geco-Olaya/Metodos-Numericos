```python
import math
import matplotlib.pyplot as plt 
from tabulate import tabulate 
%matplotlib inline

valor=[]
fun=[] 
Error=[] 
tabla1=[] 
raices=[] 
tabla2=[]
xv=14.7802
m=68.1
g=9.8
v=40.0
t=10.0
xl=12.0 
xu=16.0
xr=0.0
fa=1.0
fb=1.0

print "con c,  f(c)= \n"
for k in range(1,24):
    por=k*1
    fc=((g*m)/por)*(1-math.pow(math.e,-(por/68.1)*10.0))-v
    if(por>0 and por%4==0):
        valor.append(por)
        fun.append(fc)
        tabla1.append([por, fc])



def modo(num):
    fc=((g*m)/num)*(1-math.exp(-(num/68.1)*10.0))-v
    return fc
    
while ((fa*fb)!=0):
    xr=(xl+xu)/2
    fa=modo(xl)
    fb=modo(xr)
    erpv=math.fabs(((xv-xr)/xv)*100)
    tabla2.append([xr, str(erpv)+" %"])
    if ((fa*fb)<0):
        xu=xr
    if((fa*fb)>0):
         xl=xr   

print tabulate(tabla2, headers=["Resultado Raices", "Error"], tablefmt="fancy_grid")
print tabulate(tabla1, headers=["Valores de c", "f(c)"], tablefmt="fancy_grid")

plt.title("Valores")
plt.axis([0, 24, -10, 40])
plt.ylabel('f(c)')
plt.xlabel('c')
plt.axhline(0, color="red")
plt.plot(valor, fun, 'm-', marker='o')
plt.plot(xv,0, 'cs', label='Valor Verdadero=14.7802')
plt.legend(loc = 'upper right', numpoints = 1)
plt.show()
```
