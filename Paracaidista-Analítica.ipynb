```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

g=9.8
m=68.1
c=12.5
cont=0
tp=[]
vl=[]

for t in range(0,100):
    tp.append(t)
    v=((g*m)/c)*(1-(math.exp((-c/m)*t)))
    vl.append(v)
    print "En el tiempo: "+str(t) + " la velocidad es:" + str(v)
    
plt.plot([tp[0:]], [vl[0:]], 'o')
plt.ylabel('Velocidad (m/s)')
plt.xlabel('Tiempo (s)')
plt.axis([0, 100, 0, 100])
plt.show()
```
